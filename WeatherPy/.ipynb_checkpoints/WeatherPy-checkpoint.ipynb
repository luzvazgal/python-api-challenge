```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)
        
# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#endpoint = f"http://api.openweathermap.org/data/2.5/weather?q=queretaro&appid={weather_api_key}&units=imperial"
#response=requests.get(endpoint).json()
#response
```

```python
countries = []
latitudes = []
longitudes = []
datetimes =[]
temperatures = []
humidities = []
cloudiness = []
wind_speeds = []
city_ids = []
city_names = []


print ("Beginning Data Retrieval")
print ("-----------------------------")

for city in cities:   
    endpoint = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={weather_api_key}&units=imperial"
    
    try:
        city_data = requests.get(endpoint).json()
    
        #print(endpoint)
        print(f"Processing city: City ID: {city_data['id']} City Name: {city_data['name']}")
        #print(city_data)
        countries.append(city_data["sys"]["country"])
        latitudes.append(city_data["coord"]["lat"])
        longitudes.append(city_data["coord"]["lon"])
        datetimes.append(city_data["dt"])
        temperatures.append(city_data["main"]["temp_max"])
        humidities.append(city_data["main"]["humidity"])
        cloudiness.append(city_data["clouds"]["all"])
        wind_speeds.append(city_data["wind"]["speed"])
        city_ids.append(city_data['id'])
        city_names.append(city_data['name'])
    except:
        print (f"{city} not found... Skipping")
        continue
        

```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#Creating DataFrame

cities_df = pd.DataFrame({
        "City Id": city_ids,
        "City Name": city_names,
        "Country": countries,
        "Latitude": latitudes,
        "Longitude": longitudes,
        "Datetime": datetimes,
        "Max Temp (F)": temperatures,
        "Humidity (%)": humidities,
        "Cloudiness (%)": cloudiness,
        "Wind Speed": wind_speeds
}
)
cities_df
```

```python
#Writing to CSV file
file_path = "../output_data/cities_luz.csv"
cities_df.to_csv(file_path, sep=',')
```

## Linear Regression

```python
# OPTIONAL: Create a function to create Linear Regression plots
#Splitting cities between Northern and Southern cities

#Northern cities have Latituted value > 0
northern_cities_df = cities_df.query('Latitude>0').copy()

#Southern cities have Latituted value < 0
southern_cities_df = cities_df.query('Latitude<0').copy()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
#Creating scatter plots for each metrics' duple using Linear Regression
#Northern Hemisphere - Temperature (F) vs. Latitude scatter diagram
#Linear Regression calculation
(slope, intercept, rvalue, pvalue, stderr) = linregress(northern_cities_df["Latitude"], northern_cities_df["Max Temp (F)"])

# Get regression values
# Y (predict) = m(slope)*x(Latitude val) + b(intercept)
temp_predict = slope * northern_cities_df["Latitude"] + intercept

#Adding Temperature prediction column to DataFrame
northern_cities_df['Max Temp Predict (F)'] = temp_predict

# Create line equation string
north_temp_line_eq = f"y = {round(slope,2)}x + {round(intercept,2)}"
north_temp_line_eq

#Printing: Northern Hemisphere - Temperature (F) vs. Latitude scatter diagram
northern_cities_df.plot(x='Latitude', y='Max Temp (F)', kind='scatter', marker='o',  s=30, edgecolors='#000099')

#Printing Linear Regression line 
plt.plot(northern_cities_df["Latitude"], northern_cities_df['Max Temp Predict (F)'], c='red')

# Label plot and annotate the line equation
plt.annotate(north_temp_line_eq, (10,20), c='red')
plt.savefig('../output_data/lat_maxTemp_north.png')
plt.show()
```

```python
print(f"The r-squared is {rvalue**2}")
```

## Conclusions

Latitud vs Max Temperature

The curve seems to be a normal distribution which demonstrates "the closer to the Equator, the higher the temperature". However, in Northern hemisphere, the difference among each lat grade is higher than in Southern hemisphere. For example, at 60N, temperatyre goes from20 to 40F, while in Southern is 40F on average at -60S

Despite it,  there's more population in Northern hemisphere, but why?. According to regression linear chart, it's observed temperature prediction is almost directly proportional  to latitude, so it's easier to predict a temperature there than Southern Hemisphere, where its slope is almost the half.


There's no relationship between Lat and cloudiness and wind speed

In Northern Hemisphere, there's more humidity level in any latitude. Hurricanes may be the caused of this behavior (you will notice there are no hurricanes in Southern region), bringing water to all places.

Regarding wind speed, it is constant across all Northern Hemisphere, so there's no correlation with latitude. In Southern side, correlation is more notorious: the closer to Equator, the less wind speed.
